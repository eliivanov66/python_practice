```python
# -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30

# Определить корни

# Найти интервалы, на которых функция возрастает

# Найти интервалы, на которых функция убывает

# Построить график

# Вычислить вершину

# Определить промежутки, на котором f > 0

# Определить промежутки, на котором f < 0

import sympy
from sympy import log as log
from sympy import sin as sin
from sympy import cos as cos
from sympy import plot as plot
from sympy import solve as solve
from sympy import solveset as solveset
from sympy import Interval as interval
```

```python
# исходные функции
x = sympy.Symbol('x')
f_x = -12 * x**4 * sin(cos(x)) - 18 * x**3 + 5 * x**2 + 10 * x - 30
diff_f_x = f_x.diff()

print(f"Исходная функция: {f_x}")
print(f"Дифференциал функции: {diff_f_x}") 
```

```python
# функции для перебора
def my_f_x(arg_x):
    return -12 * arg_x**4 * sin(cos(arg_x)) - 18 * arg_x**3 + 5 * arg_x**2 + 10 * arg_x - 30

def my_diff_f_x(arg_x):
    return 12 * arg_x**4 * sin(arg_x) * cos(cos(arg_x)) - 48 * arg_x**3 * sin(cos(arg_x)) - 54 * arg_x**2 + 10 * arg_x + 10

def root_finder(arg_func, arg_lim_min, arg_lim_max, arg_step, arg_root):
    temp_my_x = []
    temp_my_f_x = []
    # поиск корня в диапазоне (-11.0, -7.5)
    my_x = arg_lim_min
    while  (my_x >= arg_lim_min) and  (my_x <= arg_lim_max):
        # промежуточные вычисления
        temp_my_x.append(my_x)
        temp_my_f_x.append(arg_func(my_x))
        # поиск перехода через 0 оси Y
        if len(temp_my_f_x) > 2:
            if temp_my_f_x[len(temp_my_f_x) - 2] * temp_my_f_x[len(temp_my_f_x) - 1] < 0 : 
                temp_root = (temp_my_x[len(temp_my_f_x) - 2] + temp_my_x[len(temp_my_f_x) - 1]) / 2
                temp_root = round(temp_root, int(log(0.1/arg_step)))
                arg_root.append(temp_root)
                break
        my_x += arg_step
    return arg_root
```

```python
# Построить график функции
print("График функции")
plot(f_x, xlim = (-12, 12), ylim = (-1000,1000))
print("График производной от функции")
plot(diff_f_x, xlim = (-12, 12), ylim = (-1000,1000))
```

```python
# поиск корней из графика для функции

# поиск корней перебором исходя из графика

f_x_roots = []
# поиск корня 1 в диапазоне (-11.0, -7.5)
f_x_result = root_finder(my_f_x, -11.0, -7.5, 0.0001, f_x_roots)
# поиск корня 2 в диапазоне (-7.5, -4.0)
f_x_result = root_finder(my_f_x, -7.5, -4.0, 0.0001, f_x_roots)
# поиск корня 3 в диапазоне (-4, 0)
f_x_result = root_finder(my_f_x, -4.0, 0.0, 0.0001, f_x_roots)
# поиск корня 4 в диапазоне (0.0, 4.0)
f_x_result = root_finder(my_f_x, 0.0, 4.0, 0.0001, f_x_roots)
# поиск корня 5 в диапазоне (4.0, 5.0)
f_x_result = root_finder(my_f_x, 4.0, 5.0, 0.0001, f_x_roots)
# поиск корня 6 в диапазоне (5.0, 10.0)
f_x_result = root_finder(my_f_x, 5.0, 10.0, 0.0001, f_x_roots)

print(f"Корни функции равны: {f_x_result}")
# Корни функции равны: [-7.65065, -5.02685, -1.33895, 2.27305, 4.38355, 8.03515]
```

```python
# поиск корней из графика для производной от функции

# поиск корней перебором исходя из графика

diff_f_x_roots = []
# поиск корня 1 в диапазоне (-12.0, -9.0)
diff_f_x_result = root_finder(my_diff_f_x, -12.0, -9.0, 0.0001, diff_f_x_roots)
# поиск корня 2 в диапазоне (-9.0, -5.0)
diff_f_x_result = root_finder(my_diff_f_x, -9.0, -5.0, 0.0001, diff_f_x_roots)
# поиск корня 3 в диапазоне (-5.0, -2.0)
diff_f_x_result = root_finder(my_diff_f_x, -5.0, -2.0, 0.0001, diff_f_x_roots)
# поиск корня 4 в диапазоне (-2.0, 0.0)
diff_f_x_result = root_finder(my_diff_f_x, -2.0, 0.0, 0.0001, diff_f_x_roots)
# поиск корня 5 в диапазоне (0.0, 1.0)
diff_f_x_result = root_finder(my_diff_f_x, 0.0, 1.0, 0.0001, diff_f_x_roots)
# поиск корня 6 в диапазоне (1.0, 2.5)
diff_f_x_result = root_finder(my_diff_f_x, 1.0, 2.5, 0.0001, diff_f_x_roots)
# поиск корня 7 в диапазоне (2.5, 5.0)
diff_f_x_result = root_finder(my_diff_f_x, 2.5, 5.0, 0.0001, diff_f_x_roots)
# поиск корня 8 в диапазоне (5.0, 9.0)
diff_f_x_result = root_finder(my_diff_f_x, 5.0, 9.0, 0.0001, diff_f_x_roots)
# поиск корня 9 в диапазоне (9.0, 12.0)
diff_f_x_result = root_finder(my_diff_f_x, 9.0, 12.0, 0.0001, diff_f_x_roots)

print(f"Корни производной от функции равны: {diff_f_x_result}")
# Корни производной от функции равны: [-9.97895, -6.83135, -4.16775, -0.39265, 0.45455, 1.70065, 3.81935, 7.00105, 9.87715]
```

```python
# возрастание или убывание функции

# возьмём значения из каждого диапазона и подставим в производную функции
way_check = [] # значения из каждого из диапазонов
way_diap = []

# подготовка значений из диапазонов корней производной функции
way_check.append(diff_f_x_result[0] - 1.0)
way_diap.append(f"(- бесконечность, {diff_f_x_result[0]})")
for i in range(1, len(diff_f_x_result)):
    way_check.append( (diff_f_x_result[i - 1] + diff_f_x_result[i])/2 )
    way_diap.append(f"({diff_f_x_result[i - 1]}, {diff_f_x_result[i]})")
way_check.append(diff_f_x_result[-1] + 1.0)
way_diap.append(f"({diff_f_x_result[-1]}, + бесконечность)")

way_increase = [] # диапазоны возрастания функции
way_decrease = [] # диапазоны убывания функции

for i in range(len(way_check)):
    temp = my_diff_f_x(way_check[i])
    if temp > 0:
        way_increase.append(way_diap[i])
    if temp < 0:
        way_decrease.append(way_diap[i])

print(f"Диапазоны возрастания функции: {way_increase}")
print(f"Диапазоны убывания функции: {way_decrease}")

# Диапазоны возрастания функции: ['(- бесконечность, -9.97895)', '(-6.83135, -4.16775)', '(-0.39265, 0.45455)', '(1.70065, 3.81935)', '(7.00105, 9.87715)']
# Диапазоны убывания функции: ['(-9.97895, -6.83135)', '(-4.16775, -0.39265)', '(0.45455, 1.70065)', '(3.81935, 7.00105)', '(9.87715, + бесконечность)']
```

```python
# найти вершины функции

# подставляем значения корней производной от фунции в фунцию

edges = []

for root in diff_f_x_result:
    edges.append(my_f_x(root))
    
print(f"Вершины функции : {edges}")
# Вершины функции : [107678.041643205, -13820.5349212159, 3111.36323513570, -32.2935799230767, -26.5127223815734, -74.1062927952254, 872.255767657110, -25610.5096752122, 72634.6640655902]
```
