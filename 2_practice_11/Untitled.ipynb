```python
import sympy
```

```python
x = sympy.Symbol('x')
print(type(x))
```

```python
f_x = 12 + x - 16 * x + 23 * x ** 2 - 10 * x ** 2
print(f_x)
```

```python
for root in sympy.solve(f_x, x):
    print(root.evalf())
```

```python
diff_f_x = f_x.diff()
print(sympy.solve(diff_f_x, x))
```

```python
sympy.plotting.plot(f_x)
```

```python
sympy.plotting.plot(diff_f_x)
```

```python
sympy.plotting.plot(1/x, xlim = (-2,2))
```
